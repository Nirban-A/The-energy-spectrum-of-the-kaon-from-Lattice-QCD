```python
import numpy as np
import matplotlib.pyplot as plt
from functions import read_files
from functions import jk_2pt_energy as jk
from functions import platue_fit
```

Different Momentum configurations for p_x p_y    

$\sqrt{(p_x^2+p_y^2)}$
- $\sqrt{0}$  -> (0 0)
- $\sqrt{1}$  -> (-1 0), (+1,0), (0,-1), (0, +1)
- $\sqrt{2}$ -> (-1 -1), (+1 -1), (-1 +1), (+1 +1)
- $\sqrt{4}$  -> (2 +0), (+2 +0), (+0 -2), (+0 +2)
- $\sqrt{5}$  -> (-2 -1), (+2 -1), (-2 +1), (+2 +1), (-1 -2), (+1 -2), (-1 +2), (+1 +2)
- $\sqrt{8}$  -> (-2 -2), (+2 -2), (-2 +2), (+2 +2)
- $\sqrt{9}$  -> (-3 +0), (+3 +0), (+0 -3), (+0 +3)
- $\sqrt{10}$ -> (-3 -1), (+3 -1), (-3 +1), (+3 +1), (-1 -3), (+1 -3), (-1 +3), (+1 +3)
- $\sqrt{16}$ -> (-4 +0), (+4 +0), (+0 -4), (+0 +4)

## $\sqrt{p_x^2+p_y^2} = \sqrt{1}$  
$\sqrt{p_x^2+p_y^2+p_z^2} = \sqrt{1}$

```python
momentum=1
raw_data_p0=read_files(base_dir_p0,momentum)
```

```python
(Energy_p0,Energy_error_p0, E_bins_p0)=jk(raw_data_p0)
```

```python
plt.title("Effective Energy plot for momentum ($p_z$) 0 (\u00B1 0.0 GeV)\n($p_x^2+p_y^2=1$)")
plt.errorbar(np.arange(len(Energy_p0)),Energy_p0,yerr=Energy_error_p0,fmt='o',ecolor='r', capsize=5, label="Energy with error")
plt.legend(loc="upper right")
plt.ylim(0.2,0.6)
plt.xlabel("t/a")
plt.ylabel("aE")
plt.show()
```

### Platue fitting for range [12,17)

```python
low=9
high=19
(platue_p0, platue_error_p0) = platue_fit(E_bins_p0,Energy_error_p0,low,high)
print(platue_p0, platue_error_p0)
```

```python
n=32
plt.title("Effective Energy plot for momentum ($p_z$) 0 (\u00B1 0.0 GeV)\n($p_x^2+p_y^2=1$)")
plt.errorbar(np.arange(len(Energy_p0[:n])),Energy_p0[:n],yerr=Energy_error_p0[:n],fmt='o', ecolor='r', capsize=5,label="Energy with error")
plt.hlines(platue_p0,low,high,color='k', label="Patue fit")
plt.legend(loc="upper right")
plt.xlabel("t/a")
plt.ylabel("$aE$")
plt.ylim(0.2,0.6)
plt.show()
```
