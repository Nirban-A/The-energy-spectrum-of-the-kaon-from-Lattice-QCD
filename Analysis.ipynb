```python
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
from functions import read_files
from functions import jk_2pt_energy as jk
from functions import jackknife_1D
from functions import platue_function
from functions import platue_fit
```

# Kaon Momentum 0

```python
base_dir_p0 = 'kaon_P0'
raw_data_p0=read_files(base_dir_p0)
```

```python
(Energy_p0,Energy_error_p0, E_bins_p0)=jk(raw_data_p0)
```

```python
plt.title("Energy ")
plt.errorbar(np.arange(len(Energy_p0)),Energy_p0,yerr=Energy_error_p0,fmt='o',ecolor='r', capsize=5)
plt.ylim(0,0.5)
plt.xlabel("t/a")
plt.ylabel("Energy")
plt.show()
```

### Platue fitting for range [8,20)

```python
low=8
high=20
(platue_p0, platue_error_p0) = platue_fit(E_bins_p0,Energy_error_p0,low,high)
print(platue_p0, platue_error_p0)
```

```python
n=32
plt.title("Energy momentum 0")
plt.errorbar(np.arange(len(Energy_p0[:n])),Energy_p0[:n],yerr=Energy_error_p0[:n],fmt='o', capsize=5)
plt.hlines(platue_p0,low,high,color='r')

plt.xlabel("t/a")
plt.ylabel("$aE$")
plt.ylim(0,0.5)
plt.show()
```
