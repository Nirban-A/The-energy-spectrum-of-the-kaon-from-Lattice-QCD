# Dispersion Graph

```python
platue_values=np.array([platue_p0,platue_p1,platue_p2,platue_p3,platue_p4,platue_p5])
platue_errors=np.array([platue_error_p0,platue_error_p1,platue_error_p2,platue_error_p3,platue_error_p4,platue_error_p5])
```

```python
print(platue_values)
print(platue_p5)
print(platue_error_p5)
```

```python
momentum_values=np.array([0,1,2,3,4,5])
```

```python
#Theoretical Values
ap=lambda n: (2*np.pi/32)*n
am=0.250
aE2_theo=[ am**2+ap(n)**2 for n in momentum_values]
```

```python
np.sqrt(aE2_theo)
```

```python
square_Error=2*platue_values*platue_errors
```

```python
square_Error
```

```python
plt.title("Dispersion Plot")
plt.errorbar(momentum_values**2,platue_values**2,yerr=square_Error,fmt='o', ecolor='r', label="Energy value from Lattice QCD", capsize=5)
plt.plot(momentum_values**2,aE2_theo,label="Theoretical Prediction")
plt.legend(loc="upper left")
plt.xlabel("$(ap)^2$")
plt.ylabel("$(aE)^2$")
plt.plot()
```
